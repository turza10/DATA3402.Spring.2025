1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
class Matrix:
    def __init__(self, n=None, m=None, values=None):
        if values is not None:
            # Ensure all rows are of equal length
            if not all(len(row) == len(values[0]) for row in values):
                raise ValueError("All rows must have the same number of columns")
            self.data = [row[:] for row in values]  # Deep copy
        elif n is not None and m is not None:
            self.data = [[0.0 for _ in range(m)] for _ in range(n)]
        else:
            raise ValueError("Must provide either dimensions (n, m) or values")

    def __getitem__(self, key):
        if isinstance(key, tuple):  
            i, j = key
            return self.data[i][j]
        return self.data[key] 

    def __setitem__(self, key, value):
        if isinstance(key, tuple):  
            i, j = key
            self.data[i][j] = value
        else:
            self.data[key] = value  

    def __eq__(self, other):
        return isinstance(other, Matrix) and self.data == other.data

    def assign(self, other):
        if isinstance(other, Matrix):
            if self.shape() != other.shape():
                raise ValueError("Matrix sizes do not match")
            self.data = [row[:] for row in other.data]
        elif isinstance(other, list):
            if not all(len(row) == len(other[0]) for row in other):
                raise ValueError("All rows in list must be of equal length")
            if self.shape() != (len(other), len(other[0])):
                raise ValueError("Matrix sizes do not match")
            self.data = [row[:] for row in other]
        else:
            raise TypeError("Assignment requires a Matrix or list of lists")

    def shape(self):
        return (len(self.data), len(self.data[0]) if self.data else 0)
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n, m, c):
    return Matrix(values=[[float(c) for _ in range(m)] for _ in range(n)])

def zeros(n, m):
    return constant(n, m, 0)

def ones(n, m):
    return constant(n, m, 1)

def eye(n):
    return Matrix(values=[[1.0 if i == j else 0.0 for j in range(n)] for i in range(n)])


print("3x2 matrix of 7s:")
print(constant(3, 2, 7).to_list())
print("2x2 zero matrix:")
print(zeros(2, 2).to_list())

print("2x3 ones matrix:")
print(ones(2, 3).to_list())

print("4x4 identity matrix:")
print(eye(4).to_list())
```

```python

print("=== Demonstrating Matrix Properties ===")

A = Matrix(values=[[1, 2], [3, 4]])
B = Matrix(values=[[0, 1], [1, 0]])
C = Matrix(values=[[2, 0], [1, 2]])
I = eye(2)


assoc_lhs = (A * B) * C
assoc_rhs = A * (B * C)
print("\nAssociativity Check: (A * B) * C == A * (B * C)")
print("Left Side:\n", assoc_lhs.to_list())
print("Right Side:\n", assoc_rhs.to_list())
print("Result:", "PASS" if assoc_lhs == assoc_rhs else "FAIL")

# 2. Distributivity
dist_lhs = A * (B + C)
dist_rhs = (A * B) + (A * C)
print("\nDistributivity Check: A * (B + C) == A * B + A * C")
print("Left Side:\n", dist_lhs.to_list())
print("Right Side:\n", dist_rhs.to_list())
print("Result:", "PASS" if dist_lhs == dist_rhs else "FAIL")

# 3. Non-Commutativity
AB = A * B
BA = B * A
print("\nCommutativity Check: A * B == B * A?")
print("A * B:\n", AB.to_list())
print("B * A:\n", BA.to_list())
print("Result:", "FAIL (Expected)" if AB != BA else "Unexpected PASS")

# 4. Identity Matrix
AI = A * I
print("\nIdentity Matrix Check: A * I == A")
print("A * I:\n", AI.to_list())
print("A:\n", A.to_list())
print("Result:", "PASS" if AI == A else "FAIL")
```
